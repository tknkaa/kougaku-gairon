```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
file_path = "/content/drive/MyDrive/工学概論/SSDSE-C-2023.csv"

data = pd.read_csv(file_path, encoding = "shift-jis")
```

```python
selected_columns = ['世帯人員', 'ぎょうざ', 'サラダ', 'すいか', 'まんじゅう', 'ビール']
selected_df = data[selected_columns]

mean_values = selected_df.mean()
std_dev_values = selected_df.std()

print("平均:")
print(mean_values)

print("\n標準偏差:")
print(std_dev_values)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
data_standardized = data.copy()
```

```python
numeric_columns = data_standardized.select_dtypes(include=['float64', 'int64']).columns
data_standardized[numeric_columns] = scaler.fit_transform(data_standardized[numeric_columns])
```

```python
data_standardized.head()
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
results = smf.ols('ビール ~ 世帯人員 + ぎょうざ + サラダ + すいか + まんじゅう', data=data_standardized).fit()
print(results.summary())
```
